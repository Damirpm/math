## Подключение библиотек. Предварительная настройка

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
```

## Импорт данных

```python
data = pd.read_excel('542-data.xlsx')
print(data)

x = data['$I$, А']
y = data['$N - N_s$, 1/c']
yerr = data['$\sigma_N$, 1/с']
```

```python
f, ax = plt.subplots()
f.suptitle(r'Спектр $\beta$ - распада')

#График
ax.plot(x, y, 'bo', label = r'Спектр $\beta$ - распада')

#Погрешность
plt.errorbar(x, y, yerr=yerr, fmt='bo', ecolor='b', elinewidth=2, capsize=5, capthick=2, alpha = 0.4)

#Подписи осей
plt.xlabel("I, А")
plt.ylabel(r'$N - N_s$, 1/с')

#Легенда
ax.legend(loc = 'best')

#Сетка
plt.grid(True)

plt.show()
```
