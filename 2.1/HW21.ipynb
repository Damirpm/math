# Задача 1

$$
    L = \sum_{i=1}^{l} (y_i - \hat{y})^2 \rightarrow \min_{\hat{y}}
$$

Функция потерь выпукла вниз, соответственно имеет минимум в точке, в которой первая производная по $\hat{y}$ равна 0. Пусть у нас есть выборка из $l$ элементов, тогда

$$
    L' = \sum_{i=1}^{l} 2(\hat{y} - y_i) = l \hat{y} - \sum_{i=1}^{l} y_i = 0 \quad \Rightarrow \quad \hat{y} = \frac{1}{l} \sum_{i=1}^{l} y_i = \ <y>
$$

# Задача 3

Покажем, что прямая, построенная по МНК, всегда проходит через точку $(\bar{x}, \bar{y})$. Из МНК можем найти коэффициенты прямой

$$
    \hat{w_1} = \frac{\hat{\text{cov}}(x, y)}{\hat{\text{var}} \, x}, \qquad \hat{w_0} = \bar{y} - \hat{w_1} \bar{x},
$$

Тогда уравнение прямой МНК

$$
    y = \hat{w_1} x + \hat{w_0} = \hat{w_1} x + \bar{y} - \hat{w_1} \bar{x} \quad \Rightarrow \quad y(\bar{x}) = \hat{w_1} \bar{x} + \bar{y} - \hat{w_1} \bar{x} = \bar{y}
$$

То есть прямая, построенная по МНК, всегда проходит через точку $(\bar{x}, \bar{y})$.

```python
df = sns.load_dataset("anscombe")
d = np.delete(np.array(df), 0, 1)

sns.lmplot(
    data=df, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)

x = np.reshape(d[:, 0], (4, 11))
y = np.reshape(d[:, 1], (4, 11))

var_x = np.mean((x - np.mean(x, axis = 1, keepdims = True)) ** 2, axis = 1)
var_y = np.mean((y - np.mean(y, axis = 1, keepdims = True)) ** 2, axis = 1)
cov_xy = np.mean((x - np.mean(x, axis = 1, keepdims = True)) * (y - np.mean(y, axis = 1, keepdims = True)), axis = 1)

print(' var_x: ', var_x, '\n','var_y: ',  var_y, '\n cov_xy: ', cov_xy)
```

$$
    \frac{∂ Z}{∂ \omega} = 2 \omega^T + 2 \lambda X = 0 \ \Rightarrow \ \omega^T = - \lambda X \ \Rightarrow \ \omega = -X^T \lambda^T
$$

$$
    \frac{∂ Z}{∂ \lambda} = (X \omega - y)^2 = 0 \ \Rightarrow \ X \omega = y \ \Rightarrow \ -X X^T \lambda^T = y
$$

$$
    \lambda^T = -(X X^T)^{-1} y \ \Rightarrow \  \omega = X^T (X X^T)^{-1} y
$$

# Задача 7

```python
data = pd.read_excel('5101data.xlsx')
print(data)

x = data['Vr']
y = data['Vsr']
```

```python
f, ax = plt.subplots(figsize=(16, 10))
f.suptitle(r'Зависимость напряжения на катушке от напряжения на резисторе')

#График
ax.plot(x, y, 'bo', label = r'Экспериментальные данные')

#Подписи осей
plt.xlabel(r'$V_r$, мВ')
plt.ylabel(r'$V_{sr}$, мВ')

#МНК
res = stats.linregress(x, y)
plt.plot(x, res.intercept + res.slope*x, 'r', label='МНК')

#Легенда
ax.legend(loc = 'best')

#Сетка
plt.grid(True)

plt.show()
```
