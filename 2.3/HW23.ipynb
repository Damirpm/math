# Задача 1

```python
from sklearn.decomposition import PCA 
N = 100
A = np.random.random((3, 3))
A = A @ A.T

X = np.random.multivariate_normal([0,0,0], A, N)
x = X[:, 0]
y = X[:, 1]
z = X[:, 2]
fig = plt.figure(figsize=(9, 16))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, label='parametric curve')

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(X)

clrs = ("r", "g", "b")
for i in range(3):
    x0, y0, z0 = pca.components_[i];
    ax.plot([0, x0], [0, y0], [0, z0], color=clrs[i])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
data = pca.transform(X)

fig, axs = plt.subplots(1, 3, figsize=(17, 7))
subscripts = ("X", "Y", "Z")

for i in range(3):
    x = data[:, i]
    y = data[:, (i+1)%3]
    axs[i].scatter(x, y, color="b")

plt.show()
```

# Задача 4

Пусть прямая $a$ задаётся уравнением $a_xx+a_yy = 0$ (тогда напрявляющий вектор $\bf{a}$ $ = (-a_y, a_x)$). Для удобства положим $\bf{a'}$ $ = (a_x, a_y)$

$$
    \rho_\Sigma^2 = \sum_{i=1}^N \frac{(a_xx_i+a_yy_i)^2}{a_x^2 + a_y^2} 
    = \frac{N}{||\bf{a'}||^2_2}\cdot ||A \bf{a'}||^2_2 → \text{extr}_a
$$
\
$$
    \frac{∂\rho_\Sigma^2}{∂\bf{a'}} = \frac{2\bf{a'}^TA^T A\cdot ||\bf{a'}||^2 - 2||A \bf{a'}||^2 a'^T}{||\bf{a'}||^4 / N} = 0
$$
\
$$
    ||\bf{a'}||^2 \cdot A^T A a'  = ||A \bf{a'}||^2 a'
$$
\
$$
    A^T A \bf{a'}  = \frac{||A \bf{a'}||^2}{||\bf{a'}||^2} \bf{a'} = λ^2 a',
$$\
где $λ$ -- одно из сингулярных чисел матрицы a. 

$\bf{a'} \perp a$, поэтому те же рассуждения применимы и к прямой, задаваемой вектором $\bf{a'}$. Значит, функция $\rho_\Sigma^2((x, y), a)$ имеет два экстремума (максимум и минимум). 

$$
    \frac{||A \bf{a}||^2}{||\bf{a}||^2} = \sup_{\bf{a}\in ℝ^2} \frac{||A \bf{a}||^2}{||\bf{a}||^2} = ||A||^2_2 = λ_\max^2 \hspace{10mm} ⇒\hspace{10mm}  \frac{||A \bf{a'}||^2}{||\bf{a'}||^2} = λ_\min^2
$$

Поэтому, так как мы делали замену $\bf{a}$ на $\bf{a'}$, 
$$\rho^2_\min = \rho_\Sigma^2((x, y), a') = N\lambda_\min^2 = \sum_{i=1}^N \frac{(a_xx_i+a_yy_i)^2}{a_x^2 + a_y^2}$$
