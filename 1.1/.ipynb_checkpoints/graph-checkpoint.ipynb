## Подключение библиотек. Предварительная настройка

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
```

## Импорт данных

```python
data = pd.read_excel('542-data.xlsx')
print(data)

x = data['$I$, А']
y = data['$N - N_s$, 1/c']
yerr = data['$\sigma_N$, 1/с']
```

```python
f, ax = plt.subplots()
f.suptitle(r'Спектр $\beta$ - распада')

#График
ax.plot(x, y, 'bo', label = r'Спектр $\beta$ - распада')

#Погрешность
plt.errorbar(x, y, yerr=yerr, fmt='bo', ecolor='b', elinewidth=2, capsize=5, capthick=2, alpha = 0.4)

#Подписи осей
plt.xlabel("I, А")
plt.ylabel(r'$N - N_s$, 1/с')

#Легенда
ax.legend(loc = 'best')

#Сетка
plt.grid(True)

plt.show()
```

## Импорт данных

```python
data = pd.read_excel('542-data.xlsx')
print(data)

x = data['$T$, кэВ']
y = data['f']
yerr = data['$\sigma_f$, mkFermi']
```

```python
x1 = np.array([0, 800])
y1 = x1*(-0.33) + [205.714, 205.714]
```

```python
f, ax = plt.subplots()
f.suptitle('График Ферми')

#Пределы
ax.set_xlim([0, 800])
ax.set_ylim([-50, 300])

#График
ax.plot(x, y, 'bo', label = 'График Ферми')
ax.plot(x1, y1, 'r-', label = 'Экстраполяция', lw = 2)

#Линии
ax.hlines(0, -100, 1000, 'm', lw = 2)
ax.vlines(623.375, -100, 1000, 'm', lw = 2)

#Погрешность
plt.errorbar(x, y, yerr=yerr, fmt='bo', ecolor='b', elinewidth=2, capsize=5, capthick=2, alpha = 0.4)

#Подписи осей
plt.xlabel("T, кэВ")
plt.ylabel(r'f,   с/м$^{3/2}$, mkFermi')

#Легенда
ax.legend(loc = 'best')

#Сетка
plt.grid(True)

plt.show()
```
