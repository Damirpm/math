```python
from datetime import datetime

x, y1, y2 = [], [], []

for i in range(1, 100, 1):
    x += [i]
    a = np.random.randint(-5, high = 5, size= (i, i))
    b = np.random.randint(-5, high = 5, size= (i, i))
    c = np.zeros((i,i))
    start = datetime.now()
    for l in range(i):
        for m in range(i):
            for p in range(i): 
                c[l][m] += a[l][p] * b[p][m]
    t = datetime.now() - start
    y1 += [t.microseconds]

    start = datetime.now()
    c = np.dot(a, b)
    t = datetime.now() - start
    y2 += [t.microseconds]
    
f, ax = plt.subplots()
f.suptitle('Время умножения')

#График
ax.plot(x, y1, 'b-', label = 'Цикл')
ax.plot(x, y2, 'm-', label = 'Numpy')

#Подписи осей
plt.xlabel("Размер массива")
plt.ylabel('t, мкс')

#Легенда
ax.legend(loc = 'best')

#Сетка
plt.grid(True)

plt.show()
```

## Упр №3. Вигнеровский полукруг.

#1

```python
bin = 17
N = 100

x = np.array([])

def eigvals():
    M = sps.norm.rvs(size=(N, N))
    M += M.T
    n, _, _ = plt.hist(np.linalg.eigvals(M), bins=bin)
    return n

for _ in range(50):
    x = np.append(x, eigvals())

x = x.reshape((50, bin))


plt.figure()
M = np.array(sps.norm.rvs(size = N*N)).reshape((N, N))
M += M.T
y, pos, _= plt.hist(np.linalg.eigvals(M), bins=bin)

middle = (pos[:-1] + pos[1:] ) / 2

plt.errorbar(middle, y, yerr=x.std(axis=0), fmt='ro', ecolor='r', elinewidth=2, capsize = (pos[2] - pos[1])*3, capthick=2, alpha = 0.4)

plt.title('Вигнеровский полукруг')
plt.xlabel('Собственные значения')
plt.ylabel('n')
plt.grid()
plt.show()
```
